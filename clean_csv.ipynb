```python
import pandas as pd
data_path = 'Archive/national parks.csv'
df = pd.read_csv(data_path)
df
```

```python


# Split the 'name,location,long&lat' column into 'name', 'location', 'latitude', and 'longitude'
df[['name', 'location', 'lat', 'long']] = df['name,location,long&lat'].str.split(',', expand=True)
# Remove the original 'name,location,long&lat' column
df.drop('name,location,long&lat', axis=1, inplace=True)

# Display the updated DataFrame
print(df)
```

```python
# Function to convert coordinates to decimal degrees
def convert_to_decimal(coord):
    parts = coord.split('°')
    degrees = float(parts[0])
    if len(parts) > 1:
        minutes = parts[1].replace('′', '')
        if '″' in minutes:
            minutes, seconds = minutes.split('″')
        else:
            seconds = '0'
        # Ensure only numeric parts are used for conversion
        minutes_numeric = ''.join(filter(str.isdigit, minutes))
        seconds_numeric = ''.join(filter(str.isdigit, seconds))
        if minutes_numeric and seconds_numeric:  # Check if not empty
            decimal_degrees = degrees + float(minutes_numeric) / 60 + float(seconds_numeric) / 3600
        else:
            decimal_degrees = degrees
    else:
        decimal_degrees = degrees
    if 'S' in coord or 'W' in coord:
        decimal_degrees = -decimal_degrees
    return decimal_degrees

# Apply the conversion function to the DataFrame
df['lat_decimal'] = df['lat'].apply(convert_to_decimal)
df['long_decimal'] = df['long'].apply(convert_to_decimal)

# Display the DataFrame with decimal degrees
print(df)
```

```python
# Drop the 'lat' and 'long' columns
df.drop(['lat', 'long'], axis=1, inplace=True)
df.rename(columns={'lat_decimal': 'lat'}, inplace=True)
df.rename(columns={'long_decimal': 'lng'}, inplace=True)
# Display the DataFrame with decimal degrees and the 'lat' column renamed
print(df)
```

```python
df.to_csv("Archive/national_PARKS.csv", index=False)
```
